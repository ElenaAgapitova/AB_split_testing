### Практическое задание №7

На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem.

Чтобы избежать различных проблем во время проведения эксперимента, необходимо составить дизайн (план) эксперимента, определить минимальный значимый эффект, основываясь на исторических данных, либо, исходя из расчета расхода на внедрение новых фич и необходимого уровня дохода, определить количества пользователей и времени необходимого для проведения эксперимента, а также провести тест для проверки системы сплитования. После подготовки и четкого планирования запускать тест. 

Ниже представлены примерные дизайны экспериментов для двух гипотез:

#### План эксперимента №1

**Гипотеза**:

*Hypothesis №5*: 
Если сделать бесплатную пробную версию подписки премиум на один месяц для всех пользователей, то мы увеличим конверсию на 1-2% в подписку, потому что у пользователя будет возможность попробовать возможности подписки, привыкнуть к преимуществам и продолжить ей пользоваться (telegram).

В связи с тем, что это социальная сеть и люди могут взаимодействовать с друг другом и общаться, то предлагать данную услугу отдельной группе не совсем этично, поэтому необходимо провести искусственный контроль: строим модель, основываясь на имеющихся данных, и выкатываем на всех.

**Что делаем**:

Контрольная версия: предсказательная модель.

Тестовая версия: предлагаем бесплатную подписку на месяц с автоматическим списанием со следующего месяца всем пользователям.

На каких пользователях тестируем: для всех пользователей.

**Ключевые метрики для оценки эксперимента:**

*Основные*:

Paying Share - доля платящих из всей аудитории

CR - доля пользователей, которые оформили подписку


*Дополнительные*:

ARPPU - средний доход от одного платящего

Retention - показатель удержания пользователей



**Ожидаемый эффект:**

Данный эксперимент требует времени, чтобы сделать выводы, так как в конечном итоге важно не только, чтобы пользователь оформил подписку, но и продолжил ей пользоваться, ориентировочно эксперимент необходимо проводить не менее 2-х месяцев с учетом окна конверсии в платную подписку, чтобы отследить CR в платную подписку, ARPU, Retention и Paying Share.

**План действий в зависимости от результатов эксперимента:**

В результате эксперимента будут получены данные, которые необходимо проанализировать и проверить, выяснить получили ли мы статистически значимые различия или разница метрик обусловлена случайностью.

Если есть статитстические значимые различия, а именно внедрение пробного бесплатного периода улучшает наши метрики, то мы можем применить данные изменения для новых пользователей (тех, которые ни разу не оформляли данную подписку, не пробовали бесплатный период).

#### План эксперимента №2

**Гипотеза**:

*Hypothesis №3*
Если переместить фильтры по подбору техники по параметрам в левую верхнюю часть экрана для всех пользователей, то мы увеличим конверсию в заказ на 1-2%, потому что пользователю будет удобнее, привычнее и проще подобрать необходимую модель из огромного количества предложенных (citilink).

**Что делаем**:

Контрольная версия: оставляем вид сайта с фильтрами, как есть.

Тестовая версия: перемещаем фильтры по подбору техники по параметрам в левую верхнюю часть экрана

На каких пользователях тестируем: на двух репрезинтативных выборках необходимого объема: выборки для тестовой и контрольных групп. 

Исходя из уровня доверия и мощности теста (например, 95% и 80% соответственно) необходимо рассчитать необходимый объем выборки, исходя из объема и окна конверсии рассчитать время необходимое для проведения экспиремента.

Перед А/В тестом необходимо провести также А/А тест для проверки системы сплитования, чтобы убедиться, что наши группы репрезентативны.

**Ключевые метрики для оценки эксперимента:**

*Основные*:

CR - доля пользователей, которые совершили заказ


*Дополнительные*:

Time Spent - время, потраченное на поиск товара

Bounce Rate- показатель отказов



**Ожидаемый эффект:**

Увеличение количества заказов, уменьшение показателя отказов и уменьшение времени на поиск необходимого товара, что влияет и на лояльность пользователя для следующих покупок.

**План действий в зависимости от результатов эксперимента:**

В результате эксперимента будут получены данные, которые необходимо проанализировать и проверить, выяснить получили ли мы статистически значимые различия или разница метрик обусловлена случайностью.

Если есть статитстические значимые различия, а именно внесенные изменения улучшают наши метрики, то мы можем применить данные изменения для всех пользователей.
