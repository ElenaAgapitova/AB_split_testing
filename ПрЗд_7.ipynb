### Практическое задание №7

На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem.


**Шаблон дизайна эксперимента**

Гипотеза:

Если сделать бесплатную пробную версию подписки премиум на один месяц для всех пользователей, то мы увеличим конверсию на 1-2% в подписку, потому что у пользователя будет возможность попробовать возможности подписки, привыкнуть к преимуществам и продолжить ей пользоваться.

В связи с тем, что это социальная сеть и люди могут взаимодействовать с друг другом и общаться, то предлагать данную услугу отдельной группе не совсем этично, поэтому либо пользователей необходимо разделить по какому-либо критерию, например, для владельцев каналов в телеграме, таким образом провести не A/B тестирование, а квази-эксперимент, либо необходимо провести искусственный контроль: строим модель, основываясь на имеющихся данных, и выкатываем на всех. Предполагаю, что искусственный контроль в данном случае подойдет больше.

Что делаем:

Контрольная версия: предсказательная модель.

Тестовая версия: предлагаем бесплатную подписку на месяц с автоматическим списанием со следующего месяца всем пользователям.

На каких пользователях тестируем: для всех пользователей.

Ключевые метрики для оценки эксперимента:

CR - доля пользователей, которые оформили подписку
ARPPU - средний доход от одного платящего
Retention - показатель удержания пользователей
Paying Share - доля платящих из всей аудитории
Ожидаемый эффект:

Данный эксперимент требует времени, чтобы сделать выводы, так как в конечном итоге важно не только, чтобы пользователm оформил подписку, но и продолжил ей пользоваться, ориентировочно эксперимент необходимо проводить 2-3 месяца, чтобы отследить ARPU, Retention и Paying Share.

План действий в зависимости от результатов эксперимента:

В результате эксперимента будут получены данные, которые необходимо проанализировать и проверить, выяснить получили ли мы статистически значимые различия или разница метрик обусловлена случайностью.

Если есть статитстические значимые различия, а именно внедрение пробного бесплатного периода улучшает наши метрики, то мы можем применить данные изменения для новых пользователей (тех, которые ни разу не оформляли данную подписку).

**Hypothesis №3**. Если переместить фильтры по подбору техники по параметрам в левую часть экрана и разместить в начале для всех пользователей, то мы увеличим конверсию в заказ на 1-2%, потому что пользователю будет удобнее, привычнее и проще подобрать необходимую модель из огромного количества предложенных.

**Hypothesis №5**. Если сделать бесплатную пробную версию подписки премиум на один месяц для всех пользователей, то мы увеличим конверсию на 1-2% в подписку, потому что у пользователя будет возможность попробовать возможности подписки, привыкнуть к преимуществам и продолжить ей пользоваться.
