- **A/B эксперимент**: разделяет аудиторию на две случайные группы, где одна группа видит текущую версию (контрольная группа), а другая - новую версию (тестовая группа). 

- **Ухудшающий A/B эксперимент**: что-то ломаем, чтобы обнаружить зависимость изменяемого объекта на выбранную метрику. Это просто и дёшево, при грамотном подходе. 

- **MVT эксперимент**: позволяет одновременно тестировать несколько вариантов изменений в разных комбинациях. За это приходится платить большим объемом выборки и качеством теста. 

- **A/A эксперимент**: как и в A/B тестах, делим на группы и проверяем, но проверяем лишь корректность системы складывания.

- **Квази-эксперимент**: имеет похожую структуру и методологию как A/B эксперимент, но не использует случайное назначение групп. Например, мы не делим пользователей равновероятно на две группы, а используем территориальный критерий. Первая группа - люди из города A, вторая группа - люди из города B 

- **Искусственный контроль**: для формирования одной из групп используются предсказательная модель.

**Примеры**:

1. При проведении прошлого тестирования оказалось, что одни и те же люди попадали сразу в две группы. из-за этого пришлось остановить эксперимент. Что может помочь разобраться в причинах? **А/А тест**

2. На сайте продаж авто есть отдельная страница, где реализован алгоритм подбора машин. Компания хочет оптимизировать работу этого алгоритма, чтобы увеличить количество продаж на 5%, и для этого готова заказать подготовку нейросети . Разработка займет приличное количество времени и бюджета. Как понять, влияет ли этот раздел сайта на объем продаж? **Ухудшающий А/В тест**

3. Наши копирайтеры придумали новый заголовок для продающей страницы, с целью увеличения time spent на 7%. Хотелось бы узнать, как это действие повлияет на удержание пользователя и в правильном ли направлении работает отдел копирайтинга. Как это можно сделать? **А/В тест**

4. Одна соц. сеть решила сделать интересное нововведение. Добавить возможность делать групповые звонки прямо в чате всем пользователям, чтобы увеличить время, проведенное в приложении - time spent - на 5%. Как оценить эффект от такого введения? **Искуственный контроль (строим предсказательную модель на имеющихся данных, выкатываем изменения на всех и смотрим результат)**
 
5. Исследователи в области образования решили измерить эффективность новой системы образования. Для этого они имеют возможность внедрить такую систему в любом учебном заведении или даже не одном. Каждый университет может использовать только одну конкретную программу. В конце предполагается измерить уровень знаний обучающихся с помощью теста. Как узнать, эффект от нововведений? **Квази-эксперимент (критерий учебное заведение)**

6. Команда дизайнеров предложила 3 новых концепции оформления продающей страницы. Основная цель - увеличить время пребывания пользователя на этой странице. Каким образом можно выбрать подходящий вариант? **MVT эксперимент**
