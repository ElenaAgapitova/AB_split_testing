```python
from scipy.stats import ttest_1samp
import numpy as np
import pandas as pd


from scipy.stats import norm, t, kstest, shapiro
import statsmodels.api as sm
from matplotlib import pyplot as plt


import warnings
warnings.filterwarnings('ignore')
warnings.warn('DelftStack')
warnings.warn('Do not show this message')
```

```python
norm_dist = norm.rvs(size=250)
t_dist = t.rvs(size=250, df=5)
```

```python
import plotly.express as px

fig = px.histogram(norm_dist, x=norm_dist, title='norm_dist', marginal='box')
fig.show()
```

```python
fig = px.histogram(t_dist, x=t_dist, title='t_dist', marginal='box')
fig.show()
```

```python
fig = sm.qqplot(norm_dist, line='45')
plt.title('Normal Q-Q Plot for normal_dist')
plt.show();
```

```python
fig = sm.qqplot(t_dist, line='45')
plt.title('Normal Q-Q Plot for t_dist')
plt.show();
```

```python
print(kstest(norm_dist, 'norm'))
print(kstest(t_dist, 'norm'))
```

```python
print(shapiro(norm_dist))
print(shapiro(t_dist))
```

```python
print(kstest(t_dist, 't', args=(5,)))
```
