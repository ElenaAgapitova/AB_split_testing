### ICE Score

**Impact (влияние)** — демонстрирует, насколько идея положительно повлияет на ключевой показатель, который вы пытаетесь улучшить.

**Confidence (уверенность)** — демонстрирует, насколько вы уверены в оценках влияния и лёгкости реализации.

**Ease (лёгкость реализации)** — оценка того, сколько усилий и ресурсов требуется для реализации этой идеи.

*Гипотеза:*

Если сделать бесплатную пробную версию подписки премиум на один месяц для всех пользователей, то мы увеличим конверсию на 1-2% в подписку, потому что у пользователя будет возможность попробовать возможности подписки, привыкнуть к преимуществам и продолжить ей пользоваться.

Предполагаю, что искусственный контроль в данном случае подойдет больше.

*Что делаем:*

Контрольная версия: предсказательная модель.
Тестовая версия: предлагаем бесплатную подписку на месяц с автоматическим списанием со следующего месяца всем пользователям.
На каких пользователях тестируем: для всех пользователей.


*Ключевые метрики для оценки эксперимента:*

* CR - доля пользователей, которые оформили подписку
* ARPPU - средний доход от одного платящего
* Retention - показатель удержания пользователей
* Paying Share - доля платящих из всей аудитории

*Ожидаемый эффект:*

Данный эксперимент требует времени, чтобы сделать выводы, так как в конечном итоге важно не только, чтобы пользователей оформил подписку, но и продолжил ей пользоваться, ориентировочно эксперимент необходимо проводить 2-3 месяца, чтобы отследить ARPU, Retention и Paying Share.
